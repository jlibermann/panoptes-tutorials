## Simple Plot

Let's start by plotting some numbers on a figure using the `plot` function.

```python
plt.plot([1, 2, 3, 4])
#plt.show()
```

Here we have plotted the numbers 1, 2, 3 and 4 along the y axis. Python does this by default when you provide the plot() command with a single list/array, as we have done. It also automatically generates x values of the same length as our array. 
As our array has 4 numbers, but python starts with 0, python generated an x array starting from 0 with 4 numbers, so [0,1,2,3]. 

-------------------------------

Let's now plot one array on the x, and another on the y by specifying both of them in the plot() command.

```python
plt.plot([0,1,2,3], [1,2,3,4])
```

This generates the exact same plot as before. 

We could also plot the same two arrays by assigning them to variables first before passing them onto the plot() command. Let's put them in numpy arrays for ease. This enables us to edit and manipulate the arrays as we like before doing any plotting.

```python
# specify the array to plot on the x axis
x = np.array([0,1,2,3])
# specify the array to plot on the y axis
y = np.array([1,2,3,4])

# plot y vs. x
plt.plot(x,y)
```

The default plot() command convention is specifying the desired x and y array, as we have done here, one after the other in the following order: `plot(<x array>, <y array>)`. These are the first two arguments.

### Scatter plot: formatting style

We can change this line graph to a scatter plot by specifying an additional third argument. This can also format the colour of the plot. This is done the format string with a letter and symbol (from MATLAB). The default setting is `'b-'`, which stands for solid blue ('b') lines ('-'). To change this line graph style to a scatter plot, we use 'o' in the string after the letter. You can change the colours as well, from blue circle: 'bo' to red: 'ro' or even green: 'go'. You can even use triangles with '^'.


Let's see try this. At the same time, let's generate some arrays automatically and use some math.
We generate a numpy array of 10 equally spaced numbers to use as the x range using numpy's `arange` function. The y values will be an array of numbers twice that of the x range. 

```python
# arrays
x = np.arange(0, 10, 1) # note this will go from 0 to 9 in steps of 1
y = x * 2

# plotting
plt.plot(x, y, 'bo')
```

-----------------------

Let's now try plotting two different scatter plots in the same figure. This is done by repeating the same three agruments we just one after the other in the same `plot()` command. We will make them different colours to distinguish them.

```python
# plot 1 arrays
x1 = np.arange(0, 10, 1)
y1 = x1 * 2

# plot 2 arrays
x2 = np.arange(0, 10, 2)
y2 = x2 + 5

# let's plot them
plt.plot(x1, y1, 'bo', x2, y2, 'ro')
```
